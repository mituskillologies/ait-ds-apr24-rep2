### filter with variance

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vt = VarianceThreshold(1.0)
```

```python
vt.fit(df)
```

```python
vt.get_support()
```

```python
newdf = df.iloc[:, vt.get_support()]
```
