```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([12,23,34,43,49,54,58,63,74,81])
y = np.array([23,28,35,41,48,53,59,68,79,86])
z = np.array([98,91,87,79,73,67,61,54,48,44])
```

### Line plot

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x, y, marker = 'o', mfc = 'red', ms = 5,
        ls = '--', linewidth = 1, color = 'blue');

plt.grid()
# marker => <, >, ^, o, H, x 
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x, z, marker = 'o', mfc = 'black', ms = 5,
        ls = '--', linewidth = 1, color = 'red');

plt.grid()
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x, y, marker = 'o', mfc = 'red', ms = 5,
        ls = '--', linewidth = 1, color = 'blue',
        label = '2023');

plt.plot(x, z, marker = 'o', mfc = 'black', ms = 5,
        ls = '--', linewidth = 1, color = 'red',
        label = '2024');

plt.grid()
plt.legend()
```

### Subplot

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x, y, marker = 'o', mfc = 'red', ms = 5,
        ls = '--', linewidth = 1, color = 'blue');

plt.grid()

plt.subplot(2,2,2)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x, z, marker = 'o', mfc = 'black', ms = 5,
        ls = '--', linewidth = 1, color = 'red');

plt.grid()
```

### Bar plot

```python
courses = ['Java', 'Kotlin', 'Python', 'R', 'C']
students = [67, 72, 75, 40, 46]
```

```python
colours = ['cyan', 'orange', 'black', 'red', 
          'blue']
```

```python
plt.title('Bar Plot')
plt.xlabel('Course')
plt.ylabel('No. of Students')
plt.bar(courses, students, color = colours, width = 0.6,
       edgecolor = 'red', linewidth = 3, alpha = 0.7);

plt.grid()
```

### Scatter Plot

```python
plt.scatter(x, y, marker = '*')
plt.grid(ls = '--', color = 'blue')
```

### Pie Chart

```python
plt.title('Pie Chart')
plt.pie(students, labels = courses, 
        autopct = '%.2f%%', shadow = True, radius = 0.8,
       labeldistance= 1.2, explode = [0.2, 0, 0, 0, 0]);
```

```python
plt.figure(figsize= (16,9))

plt.subplot(2,2,1)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x, y, marker = 'o', mfc = 'red', ms = 5,
        ls = '--', linewidth = 1, color = 'blue',
        label = '2023');

plt.plot(x, z, marker = 'o', mfc = 'black', ms = 5,
        ls = '--', linewidth = 1, color = 'red',
        label = '2024');

plt.grid()
plt.legend()

plt.subplot(2,2,2)
plt.title('Bar Plot')
plt.xlabel('Course')
plt.ylabel('No. of Students')
plt.bar(courses, students, color = colours, width = 0.6,
       edgecolor = 'red', linewidth = 3, alpha = 0.7);

plt.grid()

plt.subplot(2,2,3)
plt.scatter(x, y, marker = '*')
plt.grid(ls = '--', color = 'blue')

plt.subplot(2,2,4)
plt.title('Pie Chart')
plt.pie(students, labels = courses, 
        autopct = '%.2f%%', shadow = True, radius = 0.8,
       labeldistance= 1.2, explode = [0.2, 0, 0, 0, 0]);

plt.savefig('MY PLOT.png')
```
